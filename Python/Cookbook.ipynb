### Debugger

We can call a Debugger using the set_trace() function from IPython.core.debugger package
When using the tracer, entering c continues the execution of the program, entering n executes the current line and jumps to the next one and entering h opens a prompt with commands accepted by the tracer. This, of course, implies that if you have c or n variables in your code, them you'll have to use 'print(c)', 'print(n)' or 'print(n)' if you want to see their assigned values

```python
from IPython.core.debugger import set_trace

def add_to_meaning_of_life(x):
    
    # the next line starts the debugger when running the function
    set_trace()
    
    answer = 42
    answer += x
    
    return answer

add_to_meaning_of_life(12)
```

```python
class Molecule:
    
    def __init__(self, mol_type, macro_micro, components = None):
        self.mol_type = mol_type
        self.macro_micro = macro_micro
        self.components = components
    
    def __repr__(self):
        return '%s, %s, %s' % (self.mol_type, self.macro_micro, self.components)
    
    def __str__(self):
        return 'This molecule is a %s, which is a %s' % (self.mol_type, self.macro_micro)
```

### zip

zip takes n number of iterables and returns list of tuples. ith element of the tuple is created using the ith element from each of the iterables.


```python
list_a = [1, 2, 3, 4, 5]
list_b = ['a', 'b', 'c', 'd', 'e']

zipped_list = zip(list_a, list_b)

print zipped_list # [(1, 'a'), (2, 'b'), (3, 'c'), (4, 'd'), (5, 'e')]
```

```python
list1, list2 = [1,2,3,6,7,8,9], [4,5,6]
for i,j in zip(list1,list2):
    print i,j
```

To unzip a list of tuples we zip(*list_of_tuples). Unzip creates a separate list.

```python
zipper_list = [(1, 'a'), (2, 'b'), (3, 'c')]
list_a, list_b = zip(*zipper_list)
print list_a # (1, 2, 3)
print list_b # ('a', 'b', 'c')
```

Zip in Python3

In Python3, zip methods returns a zip object instead of a list. This zip object is an iterator. Iterators are lazily evaluated.

Lazy evaluation, or call-by-need is an evaluation strategy which delays the evaluation of an expression until its value is needed and which also avoids repeated evaluations (Wikipedia definition).

Iterators returns only element at a time. len function cannot be used with iterators. We can loop over the zip object or the iterator to get the actual list

Consider the below example:

```python
list_a = [1, 2, 3]
list_b = [4, 5, 6]

zipped = zip(list_a, list_b) # Output: Zip Object. <zip at 0x4c10a30>

# len(zipped) # TypeError: object of type 'zip' has no len()

# zipped[0] # TypeError: 'zip' object is not subscriptable

list_c = list(zipped) #Output: [(1, 4), (2, 5), (3, 6)]

list_d = list(zipped) # Output is empty list because the above statement exhausted the zipped object
list_d
```

We convert the zip object to a list by list(zipped). After this we can use all the methods of list.

Iterators can be evaluated only one time. After that they get exhausted, hence list_d output is empty list.

### default argument

default arguments are only evaluated when a function is defined, not when it is called. this can lead to issues:

```python
def function_demonstr(item, item_list = []):
    item_list.append(item)
    print(item_list)
```

```python
function_demonstr(1)
function_demonstr(2)
function_demonstr(3)
```

note that intuitively one would expect the function to create a new empty list everytime it is called, but that is only going to happen when the function is defined. This can be solved by using a default argument as None:

```python
def function_demonstr(item, item_list = None):
    if item_list is None:
        item_list = []
    item_list.append(item)
    print(item_list)
```

```python
function_demonstr(1)
function_demonstr(2)
function_demonstr(3)
```

```python
def map_values(nfolders, matrix_dir = "", directory = "/home/" + username +\
               "/Desktop/Projeto/Analises/Modelos/Python/dynamics_outputs"):
    """
    map_values will create two arrays with the values to be plotted in the heat map.
    The function first opens the final_output.txt file for each matrix_dir and gets the
    final values of x and y for every run (every different initial values of x and y).
    These values are put in values_list and then appended to x_map_values and y_map_values.
    This is done in the 'list for'. The 'arrays fors' get the values of x_map_values and
    y_map_values and allocate then into arrays. This is necessary since the heat map function
    needs arrays as inputs.
    """
    chdir(directory + "/" + matrix_dir)
    
    global x_array, y_array, x_map_values, y_map_values, values_list, array25y, array25x
    x_array, y_array, array25y, array25x = np.empty([9,9]), np.empty([9,9]), np.empty([25,9,9]), np.empty([25,9,9])
    values_list, x_map_values, y_map_values = [], [], []
    #list_name = ""
    
    for k in sorted(listdir(getcwd())): # loops through the folders in matrix_dir
        if k != "HM": # excludes HM from the loop
            with open(getcwd() + "/" + k + "/final/final_output.txt", "r") as file: # this line opens
                # the final_output.txt for all folders inside matrix_dir
                num_lines = sum(1 for line in file)
                for j in range(0, num_lines): # this loops through each final_output.txt and
                    file.seek(0)              # appends the values in each line to values_list
                    values_list.append(file.readlines()[j])
                    
            
    counter = 0
    for i in range(1, len(values_list) + 1): # reading through values_list
        if counter == 81: # this if excludes the headings of files, which don't contain any data
            counter = 0
        else:
            counter += 1
            x_map_values.append(float("{0:.2f}".format(float(values_list[i].split("\t")[4]))))
            y_map_values.append(float("{0:.2f}".format(float(values_list[i].split("\t")[6]))))
    
    counter = 0
    for k in range(0, nfolders): # this line will loop through whatever the number of
        for i in range(0,9):     # folders inside matrix_diris
            for j in range(0,9):
                if i == 0 and j == 0: # creating an array from x_map_values
                    x_array[j,i] = x_map_values[counter]
                else:
                    x_array[j,i] = x_map_values[counter]
                    if abs(x_array[j,i] - x_array[j - 1,i - 1]) < .05:
                        x_array[j,i] = x_array[j - 1,i - 1]
                        # this if ensures that two adjacent lines will have the same value
                        # if they differ for less than .05.
                        # This is necessary for creating correct heater maps: without it, the map
                        # will differentiate really small values and this will mess up the entire map
                array25x[k][j][i] = x_array[j][i]
                counter += 1
      
    counter = 0
    for k in range(0, nfolders): # this line will loop through whatever the number of
        for i in range(0,9):     # folders inside matrix_diris
            for j in range(0,9):
                if i == 0 and j == 0: # creating an array from y_map_values
                    y_array[j,i] = y_map_values[counter]
                else:
                    y_array[j,i] = y_map_values[counter]
                    if abs(y_array[j,i] - y_array[j - 1,i - 1]) < .05:
                        y_array[j,i] = y_array[j - 1,i - 1]
                        # this if ensures that two adjacent lines will have the same value
                        # if they differ for less than .05.
                        # This is necessary for creating correct heater maps: without it, the map
                        # will differentiate really small values and this will mess up the entire map
                array25y[k][j][i] = y_array[j][i]
                counter += 1
```
